## <font color='#2A7F62'> 4. Data Cleaning <a class="anchor" id="chapter4"></a>

The `count_duplicate_rows` function was used to check the dataframe for duplicate entries by utilizing the duplicated method, which identifies duplicates, and then summarizes the total number of duplicate rows.

```python
def count_duplicate_rows(df_copy):
    """
    Count the number of duplicate rows in a DataFrame.

    This function calculates the total number of duplicate rows in the DataFrame by calling the `duplicated` method,
    which marks duplicates as `True`, and then sums these cases.

    Parameters:
    df_copy (pandas.DataFrame): The DataFrame to check for duplicates.

    Returns:
    int: The count of duplicate rows.
    """
    duplicate_count = df_copy.duplicated().sum()
    return duplicate_count
```

```python
count_duplicate_rows(df_copy)
```

**Results** :
The `count_duplicate_rows(df_copy)` function returns `0`, signifying that there are **no duplicate rows** in the dataframe. This implies that all rows in the dataframe are **unique** and that the data set does not contain any **redundant information**. A result of `0` from this function suggests that the data is **clean in terms of duplicates**.


The `clean_dataset` function is designed to preprocess a dataset by addressing missing values and removing unnecessary columns. Specifically, it first drops columns that have a high percentage of missing values or are deemed irrelevant to the analysis, such as Crop Residues, IPPU, Manure applied to Soils, Manure Management, and On-farm energy use. For the remaining columns with missing values, the function applies logical imputation strategies: it fills missing values in Savanna fires, Forest fires, and Fires in humid tropical forests with 0, assuming no fires occurred in those instances. It fills missing values in Forestland, Net Forest conversion, and Food Household Consumption with the median value of each respective column, as these are continuous variables where the median provides a robust estimate. By performing these steps, the function ensures that the dataset is cleaned and ready for further analysis, facilitating accurate and reliable insights.

```python
def clean_dataset(df_copy):
    # Columns to drop
    columns_to_drop = ['Crop Residues', 'IPPU', 'Manure applied to Soils', 'Manure Management', 'On-farm energy use']
    
    # Drop the columns
    df_copy = df_copy.drop(columns=columns_to_drop)
    
    # Fill missing values
    df_copy['Savanna fires'] = df_copy['Savanna fires'].fillna(0)
    df_copy['Forest fires'] = df_copy['Forest fires'].fillna(0)
    df_copy['Forestland'] = df_copy['Forestland'].fillna(df_copy['Forestland'].median())
    df_copy['Net Forest conversion'] = df_copy['Net Forest conversion'].fillna(df_copy['Net Forest conversion'].median())
    df_copy['Food Household Consumption'] = df_copy['Food Household Consumption'].fillna(df_copy['Food Household Consumption'].median())
    df_copy['Fires in humid tropical forests'] = df_copy['Fires in humid tropical forests'].fillna(0)
    
    return df_copy
```

```python
# ensure changes above are saved in df_copy
df_copy = clean_dataset(df_copy)
```
